```python
#(1). Write a python program which searches all the product under a particular product from www.amazon.in. The
#product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
#guitars
```

```python
# Importing Libraries
import pandas as pd
import selenium                                  
from bs4 import BeautifulSoup
from selenium import webdriver                   
from selenium.webdriver.common.by import By      

import warnings                                  
warnings.filterwarnings("ignore")

import requests
from bs4 import BeautifulSoup

def search_amazon_products(product_name):
    # Define the Amazon URL for product search
    base_url = "https://www.amazon.in/s?k="
    search_url = base_url + product_name

    # Send an HTTP GET request to the search URL
    response = requests.get(search_url)

    if response.status_code == 200:
        
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all the product elements on the page
        product_elements = soup.find_all('div', class_='s-result-item')

        if not product_elements:
            print("No products found.")
        else:
            print(f"Found {len(product_elements)} products matching '{product_name}':\n")

            for product_element in product_elements:
                # Extract the product title and price if available
                product_title = product_element.find('span', class_='a-text-normal').text.strip()
                product_price_element = product_element.find('span', class_='a-price-whole')
                product_price = product_price_element.text.strip() if product_price_element else "Price not available"

                print(f"Product Title: {product_title}")
                print(f"Product Price: {product_price}\n")

    
if __name__ == "__main__":
    product_name = input("Enter the product you want to search on Amazon: ")
    search_amazon_products(product_name)
```

```python
#(3). Write a python program to access the search bar and search button on images.google.com and scrape 10
#images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import requests

# Create a folder to store the downloaded images
if not os.path.exists('downloaded_images'):
    os.makedirs('downloaded_images')

# Function to download images from a given URL
def download_image(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as file:
            file.write(response.content)

# Define the keywords
keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']

# Initialize the Chrome WebDriver
driver = webdriver.Chrome()

# Loop through the keywords
for keyword in keywords:
    try:
        # Open Google Images in the browser
        driver.get("https://www.google.com/imghp")

        # Find the search bar element by name and enter the keyword
        search_bar = driver.find_element_by_name("q")
        search_bar.clear()
        search_bar.send_keys(keyword)
        search_bar.send_keys(Keys.RETURN)

        # Scroll down to load more images (you can adjust this based on your needs)
        for _ in range(3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        # Find all the thumbnail images on the page
        thumbnail_elements = driver.find_elements_by_css_selector(".rg_i")

        # Download the first 10 images
        for i, thumbnail_element in enumerate(thumbnail_elements[:10]):
            thumbnail_element.click()
            time.sleep(2)
            full_image_element = driver.find_element_by_css_selector(".n3VNCb img")
            img_url = full_image_element.get_attribute("src")
            download_image(os.path.join('downloaded_images', f"{keyword}_{i + 1}.jpg"), img_url)
            driver.back()
            time.sleep(2)
    except Exception as e:
        print(f"An error occurred while processing '{keyword}': {str(e)}")

# Close the WebDriver
driver.quit()
```

```python
# Importing Libraries
import pandas as pd
import selenium                                  
from bs4 import BeautifulSoup
from selenium import webdriver                   
from selenium.webdriver.common.by import By      

import warnings                                  
warnings.filterwarnings("ignore")

import time                                      
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

#Importing required exceptions which needs to be handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Let's Install automated chrome browser and connect it to web driver
driver = webdriver.Chrome(service=Service())

#maximise the window
driver.maximize_window()
#opening the website
driver.get("https://www.digit.in/")
laptop = driver.find_element(By.XPATH,'/html/body/div[2]/div/ul/li[2]')
laptop.click()

best = driver.find_element(By.XPATH,'/html/body/div[2]/div/ul/li[2]/div[2]/div/div[1]/span[4]')
best.click()

gaming =driver.find_element(By.XPATH,'/html/body/div[2]/div/ul/li[2]/div[2]/div/div[5]/div/div[2]')
gaming.click()
#extracting All the Laptop url
producturl=[i.get_attribute('data-href') for i in driver.find_elements(By.XPATH,'//span[@class="datahreflink"]')[:7]]

#Creating Emptylist to extract required details
brand=[]
model=[]
os=[]
display=[]
processor=[]
storage=[]
memory=[]
launch_date=[]
price=[]

for i in producturl:
    driver.get(i)
    time.sleep(6)
    
    #Extracting Brand
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[1]/div[3]/strong'):
            brand.append(i.text)
    except NoSuchElementException:
        brand.append('-')
    
    #Extracting Product Name
    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="heading-wraper"]/h1'):
            model.append(i.text)
    except NoSuchElementException:
        model.append('-')
        
    #Extracting Os
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[3]/div/ul/li[1]/div/p[2]'):
            os.append(i.text)
    except NoSuchElementException:
        os.append('-')
        
    #Extracting display
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[3]/div/ul/li[2]/div/p[2]/strong'):
            display.append(i.text)
    except NoSuchElementException:
        display.append('-')
        
    #Extracting Processor
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[3]/div/ul/li[3]/div/p[2]'):
            processor.append(i.text)
    except NoSuchElementException:
        processor.append('-')
        
    #Extracting memory
    try:
        items = driver.find_elements(By.XPATH, '//div[@class="specs_details "]/ul')
        item = driver.find_elements(By.XPATH, '/li[4]/div/p/strong')
        for item in items:
            memory.append(item.text.split('\n')[-1])
    except Exception:
        memory.append('-')
        
    #Extracting Release date
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[1]/div[2]/strong'):
            launch_date.append(i.text)
    except NoSuchElementException:
        launch_date.append('-')
        
    #Extracting Price
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[7]/div[2]/div[3]/div[2]/div[4]/div/h2/strong'):
            price.append(i.text)
    except NoSuchElementException:
        price.append('-')
        
driver.close()
        
print(len(brand),
      len(model),
      len(os),
      len(display),
      len(processor),
      len(memory),
      len(launch_date),
      len(price))

BestGamingLaptop=pd.DataFrame({
    "Brand":brand,
    "Model":model,
    "OS":os,
    "Display":display,
    "Processor":processor,
    "Memory":memory,
    "Launch Date":launch_date,
    "Price": price
})
```

```python
#Scrolling 1000 times

for _ in range(100):
    driver.execute_script("window.scrollBy(0,1000)")
    
    comment = []
    upVte=[]
    time=[]
    
    #Extracting Comment
    cmmt=driver.find_elements(By.XPATH,'//yt-formatted-string[@id="content-text"]')
    for i in cmmt[:500]:
        comment.append(i.text)
        
    #Extracting Up Vote
    try:
        vote=driver.find_elements(By.XPATH,'//span[@id="vote-count-middle"]')
        for i in vote[:500]:
            upVte.append(i.text)
    except NoSuchElementException:
        upVte.append('-')
        
    #Extracting time
    Days = driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
    for i in Days[:500]:
        time.append(i.text)
        
print(len(comment), len(upVte), len(time))
```
